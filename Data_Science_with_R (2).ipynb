**Logistic Regression in R**

```
Logistic Regression in R:
> Logistic regression is a statistical method used for binary classification problems. In R, you can perform logistic regression using the glm function, which stands for Generalized Linear Model. There is also the lm function, but it is typically used for linear regression, not logistic regression.

> ŷ=β0+β1x1+β2x2+.....βmxm
  ŷ : Predicted value
  β0 : Intercept
  β1x1+β2x2+.... : Coefficients

> Logistic Regression in R: Performed using the glm
 function with the family parameter set to "binomial."

> Ordinary Least Squares (OLS): Used for linear regression, minimizing the sum of squared differences.

> Maximum Likelihood: Used in logistic regression to find parameter values maximizing the likelihood of the observed data.

> Squashing Function (Sigmoid): The sigmoid function transforms linear combinations into probabilities between 0 and 1 in logistic regression.
f(x)=1/1+e^-x
```

**Multiple Linear Regression**

````r
#Implementation of multiple linear regression
View(mtcars)
data<-mtcars[, c("mpg","disp","hp","drat")]
head(data)
#Finding linear relationship among variables
pairs(data, pch=0, col="black")

#Fitting the model to the dataset
model<-lm(mpg ~ disp+hp+drat, data=data)

#The distribution of model residuals should be approximated
hist(residuals(model), col="steelblue")
summary(model)
````

````r
# Load the mtcars dataset
data(mtcars)

# Display the first few rows of the dataset
head(mtcars)

# Extract the relevant variables
data <- mtcars[, c("mpg", "disp", "hp", "drat")]

# Display the first few rows of the selected variables
head(data)

# Explore the relationships among variables
pairs(data, pch = 16, col = "blue")

# Fit the multiple linear regression model
model <- lm(mpg ~ disp + hp + drat, data = data)

# Display a summary of the regression model
summary(model)

# Predictions for a new data point (example)
new_data <- data.frame(disp = 200, hp = 100, drat = 3.5)
predicted_mpg <- predict(model, new_data)
print(predicted_mpg)
````

````r
#Predictions
#MLR
#Define the coefficients from the model output
b0<-coef(summary(model))["(Intercept)", "Estimate"]
b1<-coef(summary(model))["disp", "Estimate"]
b2<-coef(summary(model))["hp", "Estimate"]
b3<-coef(summary(model))["drat", "Estimate"]

#Use the model coefficients to predict the value for
disp=220
hp=150
drat=3

print(b0 + b1*disp + b2*hp + b3*drat)
#Conclusion:
#p value=0.000000003 which is less than 0.05 so the model is considered as goodfit
````

````r
#Logistic Regression Predictions
mydata<-read.csv("http://www.karlin.mff.cuni.cz/~pesta/prednasky/NMFM404/Data/binary.csv")
#view the first few rows of the data
head(mydata)
summary(mydata)
mydata$rank<-factor(mydata$rank)
mylogit<-glm(admit~gre+gpa+rank,data=mydata,family='binomial')
summary(mylogit)

b0<-coef(summary(mylogit))["(Intercept)", "Estimate"]
b1<-coef(summary(mylogit))["gre", "Estimate"]
b2<-coef(summary(mylogit))["gpa", "Estimate"]
b3<-coef(summary(mylogit))["rank2", "Estimate"]
b4<-coef(summary(mylogit))["rank3", "Estimate"]
b5<-coef(summary(mylogit))["rank4", "Estimate"]

gre<-23
gpa<-9
rank2<-1
rank3<-0
rank4<-0


predicted<-plogis(b0 + b1*gre + b2*gpa + b3*rank2 + b4*rank3 + b5*rank4)
print(predicted)
````

**Classification by KNN**

````r
#Classification by KNN
library(class) #loading pakcage 'class' for KNN algorithm

install.packages('caret') #You can install packages in case they are not available
library('caret') # loading package 'caret' for confusion matric for variable

diabetes<-read.csv('/content/diabetes (1).csv') #loading the data

class(diabetes$mass) #class() to know datatype of the feature mass
View(diabetes)
str(diabetes) #str() to know the structure of the data

diabetes[, 'Outcome']=factor(diabetes[, 'Outcome']) #converting the 'Outcome'

str(diabetes) #checking the changes made to 'Outcome' feature
print(diabetes$age)
````

````r
mean(diabetes$Age) #finding the mean age of patients

summary(diabetes) #gives the info about mean,median, and other descriptive measures

train = diabetes[1:500,] #building training data to train KNN

test = diabetes[501:768,] #test data to test the performance

pred_test = knn(train[,-9],test[,-9],train$Outcome,k=2) #after training on the train data we calculate the output labels for the test data for k=2

pred_test #to see the output labels

confusion =table(pred_test,test$Outcome) #table() gives the us the correct  and incorrect predictions

sum(diag(confusion))/nrow(test) #this gives us the accuracy of the model of the test data

confusionMatrix(pred_test,test$Outcome) #confusion matrix gives us the accuracy ,sensitivity and other measures which helps us to interpret the model.
````

**Clustering Model**

````r
install.packages("factoextra")
library(factoextra)
library(cluster)

#Load Data
df<-USArrests
df<-na.omit(df)
df<-scale(df)

#plot number of cluster Vs Total WCSS
fviz_nbclust(df,kmeans,method='wss')
#perform K-means clustering with optimal K value
#Make this example reproducible
set.seed(1)
#perform K-means clustering with k=4 clusters
km<-kmeans(df,centers=4,nstart=25)
#View results
km
#plot results of final K-means model
fviz_cluster(km,data=df)
#find the mean of each cluster
aggregate(USArrests,by=list(cluster=km$cluster),mean)
#add cluster assignment to original data
final_data<-cbind(USArrests,cluster=km$cluster)
#View final data
head(final_data)
````

````r
# Load necessary libraries
install.packages('factoextra')
library(factoextra)
library(cluster)

# Load the iris dataset
data(iris)

# Select the numeric columns for clustering (excluding the species column)
df <- iris[, 1:4]

# Choose the number of clusters (you may use the fviz_nbclust to determine k)
k <- 3

# Perform k-means clustering
set.seed(123)  # Set seed for reproducibility
kmeans_model <- kmeans(df, centers = k, nstart = 25)

# Visualize the clustering results
fviz_cluster(kmeans_model, data = df, geom = "point", stand = FALSE)

# Add cluster assignment to the original data
final_data <- cbind(iris, cluster = kmeans_model$cluster)

# View the first few rows of the final dataset
head(final_data)
````
